```r
txt <- readLines("F:/Data_Science/FIFA 2018 Statistics.csv")
```

```r
#Split the single vector using seperating field as ","
```

```r
fieldList <- strsplit(txt,split=",")
```

```r
#Filtering out blank spaces and replacing it with 0
```

```r
assignFurther <- function(x){
    out<- character(27)
    j <- grepl("^[[:alnum:]]",x)
    out <- ifelse(j,x[j],0)
}
standardFields <- lapply(fieldList,assignFurther)
```

```r
#Tabulating the data using data.frame
```

```r
((M <- matrix(
    unlist(standardFields),
    nrow=length(standardFields),
    byrow=TRUE
)))
fifa <- as.data.frame(M,stringasFactors=FALSE)
```
